```python
clfr_all, rand_all = [{"similar":[],"substitute":[],"like":[]} for _ in range(2)]
```

```python
for (key,value) in data.items():
    if key != 'step1':
        for k in value.keys():
            if (int(key[4:]) < 12):
                clfr_all["similar"].append(str(value[k][0]))
                clfr_all["substitute"].append(str(value[k][1]))
                clfr_all["like"].append(str(value[k][2]))
            else:
                rand_all["similar"].append(str(value[k][0]))
                rand_all["substitute"].append(str(value[k][1]))
                rand_all["like"].append(str(value[k][2]))
```

```python
num_answers = len(clfr_all["like"])
```

```python
tmp = ['strong_disagree','disagree','neutral','agree','strong_agree']
```

```python
questions = ["similar","substitute","like"]
```

```python
for q in questions:
    df = pd.DataFrame()

    df["type"] = ["classifier" for _ in range(num_answers)]+["random" for _ in range(num_answers)]

    df["answers"] =pd.concat( [df.from_dict(clfr_all)[q],df.from_dict(rand_all)[q]],ignore_index=True)

    answers = list(df["answers"])

    df["answers"] = [tmp.index(x) for x in answers]

    sns.set(style = "white",palette="muted", color_codes=True,font_scale=3)

    ax = sns.boxplot(x="answers", y="type", data=df,whis=np.inf)
    sns.stripplot(x="answers", y="type", data=df,jitter=True, size=3, color=".3", linewidth=0)
    sns.plt.xticks([0,1,2,3,4], ["strongly disagree", "disagree", "neutral", "agree", "strongly agree"], rotation='60')
    ax.set_title(q)
    
    sns.plt.show()
```
